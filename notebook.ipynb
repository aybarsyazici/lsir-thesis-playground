````python
generated_report.head()
````

````python
generated_report.iloc[0]["positions"]
````

````python
display(generated_report[generated_report["is_about_event"] == False].shape)
display(generated_report[generated_report["is_about_event"] == True].shape)
````

````python
df_report = generated_report.explode("positions").reset_index(drop=True)
# Normalize the 'positions' field into a separate dataframe
df_positions = pd.json_normalize(df_report["positions"])
df_positions.head()
````

````python
display(df_positions.shape)
display(df_positions["stakeholder"].nunique())
display(df_positions[((df_positions["country.code"] == 'none') | (df_positions["country.code"] == 'None'))])
````

````python
targets_exploded = df_positions.explode("targets").reset_index(drop=True)
df_targets = pd.json_normalize(targets_exploded["targets"])
assert len(targets_exploded) == len(df_targets)
# join the targets with the positions
df_targets = pd.concat([targets_exploded, df_targets], axis=1)
````
